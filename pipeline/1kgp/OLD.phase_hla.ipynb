# Get outputs from cross validation and phase HLA types

```r
phase_hla_types = function(gene, ids)
{
    indata          = fread(paste("pipeline/impute/cv", paste("prediction", gene, "txt", sep = "."), sep = "/"), sep = "\t", header = TRUE , data.table = FALSE)
    types           = sort(unique(indata$type))
    indata$id2phase = indata$id
    indata          = indata %>% mutate(id    =            unlist(strsplit(id2phase, ":"))[[1]] )
    indata          = indata %>% mutate(phase = as.numeric(unlist(strsplit(id2phase, ":"))[[2]]))
    
    return(indata)
}


gene = "HFE"

indata = phase_hla_types(gene, ids)
```
